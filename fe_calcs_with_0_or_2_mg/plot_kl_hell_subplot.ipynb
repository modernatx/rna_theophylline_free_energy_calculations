```python
import os
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
def plotting(dir_str,plt_name):
    # Set up the subplots
    fig, axes = plt.subplots(nrows=6, ncols=3, figsize=(12, 18))

    # Format y-axis tick labels without decimals
    y_formatter = plt.FormatStrFormatter('%d')

    # Loop through each row and column
    for i in range(6):
        for j in range(3):
            dir_str_ = dir_str[i*3+j]
            x1 = np.loadtxt(os.path.join(dir_str_,'KL_Hell_bar_values.txt'))
            # t1 = np.arange(0,len(x1),1)
            x = x1[:,0]
            kl = x1[:,1]
            hell = x1[:,2]

            # Set the tick labels and axis labels for the first column and last row
            axes[i, j].yaxis.set_major_locator(MaxNLocator(integer=True))
            # axes[i, j].yaxis.set_major_formatter(y_formatter)
            # if j == 0:
            #     # axes[i, j].set_yticklabels(np.arange(0, 5.1, 1),fontsize = 16)
            #     axes[i, j].yaxis.set_major_formatter(y_formatter)
            # # else:
            # #     axes[i, j].set_yticklabels([])
            ytt = axes[i, j].get_yticklabels()
            for label in ytt:
                label.set_fontsize(14)

            if i == 5:
                # axes[i, j].set_xlabel('Time (ns)', fontsize = 15)
                xtt = axes[i, j].get_xticklabels()
                for label in xtt:
                    label.set_fontsize(14)
                # axes[i, j].set_xticklabels(xtt, fontsize = 16)
            else:
                axes[i, j].set_xticklabels([])

            # Plot the data
            barlist = axes[i, j].bar(x, kl,linewidth=.450)
            for k in range(len(kl)):
                if kl[k] > 2:
                    barlist[k].set_color('#ff0000') # set the color to red
                    barlist[k].set_edgecolor('k')
                else:
                    barlist[k].set_color('#86bf91') # set the color to green
                    barlist[k].set_edgecolor('k')
            
            axes[i,j].annotate(r'$<D_{KL}>$=%.1f'%np.mean(kl), xy=(0.9,0.9), xycoords='axes fraction',
                              horizontalalignment='right',verticalalignment='top', fontsize = 14)
            axes[i, j].set_xlim((0, 40))
            axes[i, j].hlines(2, 0, 40,colors='k', linestyles='dashed', label='', alpha=.3, linewidth=1) 

            x1 = None
            barlist = None



    fig.text(0.5, 0.07, r'$\lambda$ Window', ha='center', fontsize=20)
    fig.text(0.04, 0.5, r'$D_{KL}$', va='center', rotation='vertical', fontsize=20)
    axes[0, 0].set_ylabel('Theophylline', fontsize = 14)
    axes[1, 0].set_ylabel('1-Methylxanthine', fontsize = 14)
    axes[2, 0].set_ylabel('3-Methylxanthine', fontsize = 14)
    axes[3, 0].set_ylabel('Hypoxanthine', fontsize = 14)
    axes[4, 0].set_ylabel('Xanthine', fontsize = 14)
    axes[5, 0].set_ylabel('Caffeine', fontsize = 14)

    axes[0, 0].set_title('rep. 1', fontsize = 14)
    axes[0, 1].set_title('rep. 2', fontsize = 14)
    axes[0, 2].set_title('rep. 3', fontsize = 14)


    # Show the plot
    plt.savefig(os.path.join('/home/arasouli/repos/alchemical/rna_small_molecule_FE/results/dU_plot/',plt_name), bbox_inches='tight')
```
