```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lig_rmsd = np.loadtxt('lig_rmsd.dat')
rna_rmsd = np.loadtxt('rna_rmsd.dat')
```

```python
def simplePlot(numpyArray, labelX, labelY, figTitle, filename, yRange=()):
       fig, ax = plt.subplots()
       x, y = numpyArray[:, 0]/5, numpyArray[:, 1]

       ax.plot(x, y)

       ax.set(xlabel=str(labelX), ylabel=str(labelY),
              title=figTitle)
       
       if yRange:
        plt.ylim(yRange)

       fig.savefig(str(filename) + ".png")
       plt.show()
```

```python
simplePlot(lig_rmsd,"Time", "RMSD", "Ligand RMSD", "lig_rmsd")
simplePlot(rna_rmsd,"Time", "RMSD", "RNA RMSD", "rna_rmsd")
```
