```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dir_list = ['2-theophylline', '3-1_methylxanthine', '4-3_methylxanthine', '5-hypoxanthine', '6-xanthine', '7-caffeine']
cond_list = ['3-55NaCl_Mg' ]
```

```python
files = []
for c in cond_list:
    for d in dir_list:
        file = d + '_' + c + '_noPBC' + '.txt'
        files.append(file)
rdfs = []       
for f in files:
    x = np.loadtxt(f, usecols=(1)) 
    rdfs.append(x)
rdf_stacked = np.column_stack(rdfs)
rdf_mean = np.mean(rdf_stacked,axis=1)
```

```python
files = []
for c in cond_list:
    for d in dir_list:
        file = d + '_' + c + '_OPC_noPBC' + '.txt'
        files.append(file)
rdfs = []       
for f in files:
    x = np.loadtxt(f, usecols=(1)) 
    rdfs.append(x)
opc_rdf_stacked = np.column_stack(rdfs)
opc_rdf_mean = np.mean(opc_rdf_stacked,axis=1)
```

```python
r = np.loadtxt('2-theophylline_3-55NaCl_Mg_noPBC.txt', usecols=(0))
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(r, rdf_mean, label='RDF Na$^+$ TIP3P, 2 Mg$^{2+}$', linewidth=2)
ax.plot(r, opc_rdf_mean, label='RDF Na$^+$ TIP3P, 2 Mg$^{2+}$', linewidth=2)
```
