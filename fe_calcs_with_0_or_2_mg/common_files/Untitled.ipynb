```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate some random time series data
np.random.seed(0)
time_series = np.random.randn(1000)

# Calculate mean and standard deviation
mean_val = np.mean(time_series)
std_val = np.std(time_series)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the time series
ax1.plot(time_series)
ax1.set_title('Time Series')

# Plot the histogram and KDE of the time series
sns.histplot(time_series, bins=30, kde=True, ax=ax2, color='skyblue')
ax2.set_title('Histogram with KDE')

# Add mean and standard deviation annotations to histogram
ax2.axvline(mean_val, color='r', linestyle='--', linewidth=1.5, label='Mean')
ax2.axvline(mean_val + std_val, color='g', linestyle='--', linewidth=1.5, label='Mean + Std')
ax2.axvline(mean_val - std_val, color='g', linestyle='--', linewidth=1.5, label='Mean - Std')
ax2.legend()

# Add labels and adjust layout
fig.suptitle('Time Series and Histogram with KDE')
plt.tight_layout()

# Show the plot
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate some random time series data
np.random.seed(0)
time_series = np.random.randn(1000)

# Create a figure with two subplots using gridspec
fig = plt.figure(figsize=(12, 6))
gs = fig.add_gridspec(1, 2, width_ratios=[2, 1])

# Plot the time series on the left
ax1 = fig.add_subplot(gs[0])
ax1.plot(time_series, color='dodgerblue', label='Time Series')
ax1.set_title('Time Series')
ax1.set_xlim(0, len(time_series))  # Set x-axis limits
ax1.tick_params(axis='both', which='both', labelsize=13)
ax1.legend()

# Plot the histogram on the right using the y-axis of the time series
ax2 = fig.add_subplot(gs[1])
counts, bins, _ = ax2.hist(time_series, bins=30, color='skyblue', orientation='horizontal', density=True)
ax2.set_ylim(ax1.get_ylim())
ax2.set_title('Histogram')

# Compute mean and std
mean = np.mean(time_series)
std = np.std(time_series)

# Plot mean line
ax2.axhline(mean, color='red', linestyle='-', linewidth=2, label='Mean')

# Plot mean ± std box
ax2.axhspan(mean - std, mean + std, facecolor='lightgreen', alpha=0.3, label='Mean ± Std')

# Add text for mean and std values
ax2.text(0.05, mean+.1, f'Mean: {mean:.1f}', color='red', ha='left', va='center')
ax2.text(0.05, mean - std, f'Std: {std:.1f}', color='darkgreen', ha='left', va='center')

# Remove y-axis tick labels from the histogram subplot
ax2.set_yticklabels([])
ax2.tick_params(axis='both', which='both', labelsize=13)

# Set labels and adjust layout
# fig.suptitle('Time Series and Histogram')
plt.tight_layout()

# Show the plot
plt.show()



```
