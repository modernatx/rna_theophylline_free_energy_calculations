```python
prmtop_1 = ['/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop']
            
prmtop_2 = ['/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop',
            '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/1-sys_prep/box.prmtop']
# List of filenames of the first set of DCD files for TIP3P
dcds_1 = ['/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/1-40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/1-40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/1-40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/1-40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/1-40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/1-40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd']

# List of filenames of the first set of DCD files for OPC
dcds_2 = ['/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/2-theophylline/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/3-1_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/4-3_methylxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/5-hypoxanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/6-xanthine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/1-rep1/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/2-rep2/2-sim_run/equ_2/equ.0.dcd',
          '/home/arasouli/repos/alchemical/rna_small_molecule_FE/7-caffeine/3-55NaCl_Mg/6-opc_40winCmplx_30winLig/3-rep3/2-sim_run/equ_2/equ.0.dcd']

# Load all the DCD files from the first set into a list of universes
universes1 = [mda.Universe(prmtop, filename) for filename, prmtop in zip(dcds_1,prmtop_1)]

# Load all the DCD files from the second set into a list of universes
universes2 = [mda.Universe(prmtop, filename) for filename, prmtop in zip(dcds_2,prmtop_2)]

# Combine all the frames from all the universes from the first set into one list
frames1 = []
for universe in universes1:
    for ts in universe.trajectory[0::200]:
        rna = universe.select_atoms("nucleicbackbone")
        frames1.append(rna.positions)
        
    
# Combine all the frames from all the universes from the second set into one list
frames2 = []
for universe in universes2:
    for ts in universe.trajectory[0::200]:
        rna = universe.select_atoms("nucleicbackbone")
        frames2.append(rna.positions)
        
        
# Combine all the frames from both sets into one list
all_frames = frames1 + frames2

# Combine all the positions from all the frames into one array
all_positions = np.concatenate(all_frames)
```

```python
# Perform PCA on the combined positions
pca = PCA(n_components=2)
pca.fit(all_positions)

# Project all the positions onto the first two principal components
pc_projections = pca.transform(all_positions)

# Divide the projections into separate arrays for each set
set1_projections = pc_projections[:len(frames1)*len(frames1[0])]
set2_projections = pc_projections[len(frames1)*len(frames1[0]):]
```

```python
# Plot the projections in the PCA space
plt.scatter(set1_projections[:, 0], set1_projections[:, 1], alpha=.1, s=1, color='red', label='TIP3P')
# plt.scatter(set2_projections[:, 0], set2_projections[:, 1], alpha=.1, s=1, color='blue', label='OPC')
plt.legend()
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("RNA Structural Differences")
plt.savefig(f"./figs/PCA_TIP3P.pdf", bbox_inches='tight')
plt.show()
```

```python
# Plot the projections in the PCA space
plt.scatter(set2_projections[:, 0], set2_projections[:, 1], alpha=.1, s=1, color='blue', label='OPC')
plt.legend()
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("RNA Structural Differences")
plt.savefig(f"./figs/PCA_OPC.pdf", bbox_inches='tight')
plt.show()
```

```python
# Plot the projections in the PCA space
plt.scatter(set1_projections[:, 0], set1_projections[:, 1], alpha=.1, s=1, color='red', label='TIP3P')
plt.scatter(set2_projections[:, 0], set2_projections[:, 1], alpha=.1, s=1, color='blue', label='OPC')
plt.legend()
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("RNA Structural Differences")
plt.savefig(f"./figs/PCA_TIP3P_OPC.pdf", bbox_inches='tight')
plt.show()
```
